```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

## Merging the datasets

```python
Final_data = pd.merge(Data ,Data2 , on = "Country Code", how = "inner")
```

```python
Final_data.head(2)
```

### Drop Unuseful Columns

```python
Final_data.drop(columns=['Country Code','Indicator Name','SpecialNotes','TableName','Unnamed: 5'],inplace=True)
```

```python
Final_data.head(2)
```

### EDA and Data Visualization

```python
# sort data based on total population for 2022
total_population_sort = Final_data.sort_values(by='2022',ascending=False , na_position ="last")
total_population_sort.head(3)
```

## get top 10 countries with the highest population for 2022 and 2000

```python
plt.figure(figsize=(20, 13))
plt.rcParams['axes.facecolor'] = "black"

# Top 10 countries in 2022
plt.subplot(2, 2, 1)
top_10_2022 = total_population_sort.nlargest(10, '2022')
sns.barplot(x='2022', y='Country Name', data=top_10_2022, palette='coolwarm')
plt.title("Top Ten countries Of Total Population (2022)")
plt.xlabel("Total population")
plt.ylabel("Country")

# Top 10 countries in 2000
plt.subplot(2, 2, 2)
top_10_2000 = total_population_sort.nlargest(10, '2000')
sns.barplot(x='2000', y='Country Name', data=top_10_2000, palette='coolwarm')
plt.title("Top Ten countries Of Total Population in 2000")
plt.xlabel("Total population")
plt.ylabel("Country")

plt.tight_layout()
plt.show()
```

## Top Ten countries of total population in year 1999 and 1989

```python
plt.figure(figsize=(20,13))

plt.subplot(2,2,1)
top_10_1999 = total_population_sort.nlargest(10,"1999")
sns.barplot(x="1999" ,y ="Country Name", data =top_10_1999 , palette="coolwarm")
plt.title("Top Ten Countries of Total Population(1999)")
plt.xlabel("Total Population")
plt.ylabel("Country")


plt.subplot(2,2,2)
top_10_1989 = total_population_sort.nlargest(10,"1989")
sns.barplot(x="1989" , y="Country Name", data=top_10_1999, palette="coolwarm")
plt.title("Top Ten Countries of Total Population(1989)")
plt.xlabel("Total Population")
plt.ylabel("Country")

plt.tight_layout()
plt.show()
```

## Top Ten countries of total population in year 1964 and 1962

```python
plt.figure(figsize=(20,13))

plt.subplot(2,2,1)
top_10_1964 = total_population_sort.nlargest(10,"1964")
sns.barplot(x="1964" ,y ="Country Name", data =top_10_1964 , palette="coolwarm")
plt.title("Top Ten Countries of Total Population(1964)")
plt.xlabel("Total Population")
plt.ylabel("Country")


plt.subplot(2,2,2)
top_10_1962 = total_population_sort.nlargest(10,"1962")
sns.barplot(x="1962" , y="Country Name", data=top_10_1962, palette="coolwarm")
plt.title("Top Ten Countries of Total Population(1962)")
plt.xlabel("Total Population")
plt.ylabel("Country")

plt.tight_layout()
plt.show()
```

## Region Wise Population in 2022 and 2015

```python
plt.figure(figsize=(16, 9))
plt.rcParams['axes.facecolor'] = "black"

# Group by region and aggregate population data
df_2022 = total_population_sort.groupby("Region")["2022"].sum()
df_2015 = total_population_sort.groupby("Region")["2015"].sum()

# plot for 2022
plt.subplot(1, 2, 1)
sns.barplot(x=df_2022, y=df_2022.index, palette='coolwarm')
plt.title("Region Wise Population in 2022")
plt.xlabel("Total population")
plt.ylabel("Region")

# plot for 2015
plt.subplot(1, 2, 2)
sns.barplot(x=df_2015, y=df_2015.index, palette='coolwarm')
plt.title("Region Wise Population in 2015")
plt.xlabel("Total population")
plt.ylabel("Region")

plt.tight_layout()
plt.show()
```

## Region wise Population in 1964 and 1960

```python
plt.figure(figsize=(16, 9))
plt.rcParams['axes.facecolor'] = "black"

# Group by region and aggregate population data
df_1964 = total_population_sort.groupby("Region")["1964"].sum()
df_1960 = total_population_sort.groupby("Region")["1960"].sum()

# plot for 1964
plt.subplot(1, 2, 1)
sns.barplot(x=df_1964, y=df_1964.index, palette='coolwarm')
plt.title("Region Wise Population in 1964")
plt.xlabel("Total population")
plt.ylabel("Region")

# plot for 1960
plt.subplot(1, 2, 2)
sns.barplot(x=df_1960, y=df_1960.index, palette='coolwarm')
plt.title("Region Wise Population in 1960")
plt.xlabel("Total population")
plt.ylabel("Region")

plt.tight_layout()
plt.show()
```

### Analyzing Global Population Trends and Gender Ratios Across Countries Over Recent Years

```python
Total_population= pd.read_csv(r"C:\Users\Mugdha dogra\Downloads\API_SP.POP.TOTL.MA.IN_DS2_en_csv_v2_459939\TOtal_male.csv")
Data = pd.read_csv(r"C:\Users\Mugdha dogra\Downloads\API_SP.POP.TOTL.MA.IN_DS2_en_csv_v2_459939\Country.csv")
Male_population = pd.merge(Total_population , Data , on ="Country Code" , how = "inner")
Male_population.head(2)
```

```python
# Sorting the data to create the visualization of top populated countries
Data = Final_data.sort_values(by ="2019" , ascending = False, na_position ="last")
Data2 = Male_population.sort_values(by="2019", ascending = False, na_position="last")

#plotting the bar graph for population in 2019
fig = plt.figure(figsize =(15,4))

# setting index the create a grouped bar chart
barwidth = 0.4
index = Data["Country Code"].head(30)
new_index = [i + barwidth for i in range(len(index))]

plt.bar(x=index, height=Data["2019"].head(30),label = "Total_Population",width=barwidth, color="b")
plt.bar(x=new_index, height=Data2["2019"].head(30),label = "Male_Population",width=barwidth, color="r")

plt.title("Male Population and Total  Population in 2019")
plt.xlabel("Country Code")
plt.ylabel("Population")
plt.xticks(rotation=70)

plt.legend()
plt.show()
```

```python
# Sorting the data to create the visualization of top populated countries
Data = Final_data.sort_values(by ="2021" , ascending = False, na_position ="last")
Data2 = Male_population.sort_values(by="2021", ascending = False, na_position="last")

#plotting the bar graph for population in 2021
fig = plt.figure(figsize =(15,4))

# setting index the create a grouped bar chart
barwidth = 0.4
index = Data["Country Code"].head(30)
new_index = [i + barwidth for i in range(len(index))]

plt.bar(x=index, height=Data["2021"].head(30),label = "Total_Population",width=barwidth, color="b")
plt.bar(x=new_index, height=Data2["2021"].head(30),label = "Male_Population",width=barwidth, color="r")

plt.title("Male Population and Total  Population in 2021")
plt.xlabel("Country Code")
plt.ylabel("Population")
plt.xticks(rotation=70)

plt.legend()
plt.show()
```
